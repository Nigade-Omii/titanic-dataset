```python
train_accuracy= []
test_accuracy = []

for i in range(1,31):
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train,y_train)
  trainac = knn.score(X_train,y_train)
  train_accuracy.append(trainac)
  testac=knn.score(X_test,y_test)
  test_accuracy.append(testac)
 
```
